```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import svm
from sklearn.svm import SVC 
le = preprocessing.LabelEncoder() 
%matplotlib inline
```

```python
data=pd.read_csv("/content/drive/MyDrive/Iris.csv")
data.head()
```

```python
col=data.columns
col
```

```python
y=data["Species"].values
data=data.drop(["Id"],axis=1)
X=data.drop(["Species"],axis=1)
y=le.fit_transform(y)
y
```

```python
X
```

```python
X["SepalLengthCm"]=X["SepalLengthCm"].astype("category")
X["PetalLengthCm"]=X["PetalLengthCm"].astype("category")
X["SepalWidthCm"]=X["SepalWidthCm"].astype("category")
X["PetalWidthCm"]=X["PetalWidthCm"].astype("category")
X["SepalLengthCm"]=X["SepalLengthCm"].cat.codes
X["PetalLengthCm"]=X["PetalLengthCm"].cat.codes
X["SepalWidthCm"]=X["SepalWidthCm"].cat.codes
X["PetalWidthCm"]=X["PetalWidthCm"].cat.codes
X
```

# **LINEAR SVM**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
lin=SVC(kernel='linear')
lin.fit(X_train, y_train)
```

```python
y_pred=lin.predict(X_test)
y_pred
```

```python
accuracy = lin.score(X_test, y_test)
accuracy
```
